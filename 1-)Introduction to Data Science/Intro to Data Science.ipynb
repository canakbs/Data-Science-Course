# Introduction to Data Science


### What is data?

Data is information about any features of any object. We produce billions of data in a day. Our heartbeat per minute is a data or our heigth is a data as well. These are some basic examples but we can push the boundries by thinking about artifical intelligence or computer vision. 

### What is data collection

We can collect our heartbeats data by writing down them for a minute or we can collect the number of cats that we saw in a day regulary. These are called data collection but <b> WHY WE COLLECT DATA?</b> 

It is reguired to have good data about any related problem to analyze the problem well or train a model to solve the problem. These means we use data to work better.

### Quality of Data

If we use bad data with a lot of null values or a lot of dismatches probably we get bad results because we can not use bad-dirty data to analyze the data or train. This is called <b>Garbage In Garbage Out</b>.

<b>How we can avoid to use bad data or dirty data?</b>

We have a lot of methodology to regain or use the bad data. Each of them has pros and cons.
For example: We can delete all of the bad data and work with the remaining. We don't spend much time by deleting the data but most of the data could be gone. The other example: We can get the <b>means</b> of continuous values if it is null. We use all of the null values but we manipulate the data to be more tend to be mean of the continuous values.

### What is data point?


Data point is data as well but we call it as a point because of that there can be multiple of dimensity of a data. For example: Our data is about humans and we collect the data about gender, height, weight, hair color etc.

Here is how a computer sees Alice and Bob in a structured way (like an Excel table):

| Name | Gender | Height (m) | Weight (kg) | Hair Color |
| :--- | :--- | :--- | :--- | :--- |
| **Alice** | Female | 1.60 | 55 | Black |
| **Bob** | Male | 1.85 | 80 | Brown |

* **Rows (Alice, Bob):** These are the **Samples** or **Data Points**.
* **Columns (Height, Weight):** These are the **Features** or **Dimensions**.

### How can we use Data Science in the Real World?

We have learned what data is. Now, let's explore what we can do with the data we have obtained and how we can utilize it.

Here are some of examples of the fields of data science:

-<b> Data Analysis & Visualization:</b> Creating charts/graphs using large datasets, Dashboarding, Data Grouping & Aggregation.

-<b> Prediction (Machine Learning):</b> Churn Analysis, Price Prediction, Sales Forecasting, Category Classification.

-<b> Computer Vision:</b> Disease Detection (MRI/X-Ray), Early Diagnosis, Object Detection, Object Counting.

-<b> NLP (Natural Language Processing):</b> Sentiment Analysis, Text Labeling, NER (Named Entity Recognition), Keyword Extraction, Text Summarization.

-<b> Generative AI & LLMs:</b> Large Language Models (GPT, Gemini), Chatbots, Virtual Voice Assistants.

### How do we solve problems? (The CRISP-DM Cycle)

We learned about data and different fields. But how do we start a project? We don't just jump into coding. We follow a path.

This path is called **CRISP-DM**. It is like a map for Data Scientists.

Cross-Industry Standard Process for Data Mining

<img src="crisp dm.png" width="400">

Here are the 6 steps of this cycle:

1.  **Business Understanding:** First, we must ask: "What is the problem?" and "What do we want to achieve?". We need to understand the goal.
2.  **Data Understanding:** We look at our data. Is it enough? Is it clean? What features do we have?
3.  **Data Preparation:** This is the "Cleaning" part. We fix the bad data (missing values, errors). **Note:** This usually takes the most time!
4.  **Modeling:** Now we use Machine Learning algorithms. We train our model here.
5.  **Evaluation:** We test our model. Did it work? Is the result good enough? If not, we go back to the beginning.
6.  **Deployment:** If the model is good, we use it in real life (like an app or a website).

**Important:** As you see in the image, this is a **cycle (circle)**. It never truly ends. We always try to make it better.
